```python
class Harness(MetaSystem.HyperSystem):
    edge_name = "level"
    
    @classmethod
    def validate(cls):
        for node in cls.relationships():
            assert type(int(node[cls.edge_name])), f"Invalid value for level: {node[cls.edge_name]}"
    
    @classmethod
    def entities_below_level(cls, level=1):
        for node in cls.relationships():
            if int(node[cls.edge_name]) < level:
                yield node
```
